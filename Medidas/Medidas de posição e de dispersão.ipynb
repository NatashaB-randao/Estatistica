#### Passo 1: Introdução
As medidas de posição e de dispersão são ferramentas essenciais na análise de dados. Elas nos ajudam a entender onde os dados estão localizados e o quão espalhados estão, respectivamente. Imagine que você está organizando uma festa e precisa saber onde a maioria dos seus amigos mora e o quão longe alguns deles podem vir. As medidas de posição e dispersão te ajudam nessa tarefa.

#### Passo 2: Definições


##### Medidas de posição:

- **Média**: É a soma de todos os valores dividida pelo número total de observações. Em outras palavras, é o valor "médio" dos dados.
- **Moda**: É o valor que aparece com mais frequência em um conjunto de dados. É útil para identificar o valor mais comum.
- **Mediana**: É o valor que está no meio de um conjunto de dados quando eles são ordenados. Se houver um número par de observações, a mediana é a média dos dois valores do meio.

##### Medidas de variabilidade:

- **Mínimo**: É o menor valor presente no conjunto de dados.
- **Máximo**: É o maior valor presente no conjunto de dados.
- **Amplitude**: É a diferença entre o valor máximo e o valor mínimo. Indica a variação total dos dados.
- **Quartis**: Os quartis dividem um conjunto de dados ordenados em quatro partes iguais, cada uma representando 25% dos dados.
- **Percentis**: Indica a porcentagem de dados que está abaixo de um determinado valor.
- **Variância**: Mede a dispersão dos dados em relação à média. É calculada como a média dos quadrados das diferenças entre cada valor e a média.
- **Desvio Padrão**: É a raiz quadrada da variância. Indica o quanto os valores do conjunto estão desviando da média.

Essas medidas são importantes para entender tanto a tendência central dos dados quanto sua dispersão, ajudando na interpretação e análise dos conjuntos de dados.

#### Passo 3: Exemplos e Cálculos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Conjunto de dados
grupos = {
    'A': [3, 4, 5, 6, 7],
    'B': [1, 3, 5, 7, 9],
    'C': [5, 5, 5, 5, 5],
    'D': [3, 5, 5, 7],
    'E': [3, 5, 5, 6, 6]
}
```

```python
# Função 

def calcular_medidas(grupo):
    medidas = {}
    medidas['Média'] = np.mean(grupo)
    medidas['Mediana'] = np.median(grupo)
    medidas['Moda'] = max(set(grupo), key=grupo.count)  # Calcula a moda
    quartis = np.percentile(grupo, [25, 50, 75])
    medidas['Primeiro Quartil'] = quartis[0]
    medidas['Segundo Quartil (Mediana)'] = quartis[1]
    medidas['Terceiro Quartil'] = quartis[2]
    medidas['Variância'] = np.var(grupo)
    medidas['Desvio Padrão'] = np.std(grupo)
    medidas['Mínimo'] = np.min(grupo)
    medidas['Máximo'] = np.max(grupo)
    medidas['Amplitude'] = np.max(grupo) - np.min(grupo)
    return medidas
```

```python
#  Exemplos e Cálculos
for grupo, dados in grupos.items():
    print(f"\nGrupo {grupo}:")
    medidas = calcular_medidas(dados)
    for medida, valor in medidas.items():
        print(f"{medida}: {valor}")
```

```python
# Definindo uma paleta de cores
cores = plt.cm.viridis(np.linspace(0, 1, len(grupos)))

# Visualização de Dados
for i, (grupo, dados) in enumerate(grupos.items()):
    # Histograma
    plt.hist(dados, bins=10, color=cores[i], label=f'Grupo {grupo}')
    plt.title('Histograma')
    plt.xlabel('Valores')
    plt.ylabel('Frequência')
    plt.legend()
    plt.show()

    # Boxplot
    plt.boxplot(dados, patch_artist=True, boxprops=dict(facecolor=cores[i]), labels=[f'Grupo {grupo}'])
    plt.title('Boxplot')
    plt.ylabel('Valores')
    plt.xlabel('Grupo')
    plt.show()
```

```python
# Conclusão
print("\nConclusão: As medidas de resumo (posição e variabilidade/dispersão) correspondem a uma síntese do conjunto de dados observados.")
print("Enquanto as medidas de posição ou de tendência central dão ideia do centro em torno do qual os dados se distribuem,")
print("as medidas de variabilidade ou dispersão revelam o quanto se dispersam.")
```
