```python
import pandas as pd
# import sqlalchemy

# Use the full path
df = pd.read_csv(r"D:\Natasha\Estatistica\Curso de Estatística\data\points_tmw - dados origem.csv")
df.head()
```

```python
# Tabela de Frequência
# pd.value_counts(df['descProduto'])

# Tabela de Frequência
freq_produto = df.groupby(["descProduto"])[["idTransacao"]].count()

freq_produto

```

```python
# Frequência Acumulada
freq_produto["Frequência Absoluta Acumulada"] = freq_produto["idTransacao"].cumsum()
freq_produto
```

```python
# Frequência Relativa
freq_produto["Frequência Relativa"] = freq_produto["idTransacao"] / freq_produto["idTransacao"].sum()
freq_produto
```

```python
# Frequência Relativa Acumulada
freq_produto["Frequência Relativa Acumulada"] = freq_produto["Frequência Relativa"].cumsum()
freq_produto
```
